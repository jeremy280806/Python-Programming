```python
# Solving Linear Equation E231-MATH6183-JJ01-02
import numpy as np

def check_diagonally_dominant(A):
    diag = np.diag(np.abs(A))
    not_diag = np.sum(np.abs(A), axis=1) - diag
    for i in range(len(diag)):
        if diag[i] < not_diag[i]:
            print("\nNot Diagonally Dominant!")
            return False
    print("\nDiagonally Dominant!")
    return True

def gauss_seidel(A, b, e=0.022, n=15):
    array_old = np.zeros(A.shape[0])
    diag = np.diag(A)
    A_mod = np.array(A, dtype=float)
    np.fill_diagonal(A_mod, 0)
    all_errors = []

    for i in range(n):
        array_new = np.array(array_old)
        for j, row in enumerate(A_mod):
            array_new[j] = (b[j] + np.dot(row, array_new)) / diag[j]

        error_rel = np.abs((array_new - array_old) / array_new) * 100
        all_errors.append(error_rel.tolist())

        print(f"Iter {i+1}: {array_new}, Error = {error_rel}")

        euclid_dist = np.sqrt(np.dot(array_new - array_old, array_new - array_old))
        if euclid_dist < e:
            print("Converged!")
            print("\nDaftar Error Relatif (%) setiap iterasi: ")
            for idx, err in enumerate(all_errors):
                print(f"Iter {idx+1}: {err}")
            return
        array_old = array_new

    print("Tidak konvergen hingga iterasi maksimum.")

# Matriks A dan b
Xs = [
    [
        [4, 2, -1],
        [1, -5, 2],
        [2, -1, -4]
    ],
    [
        [3, 4, 5],
        [-3, 7, -4],
        [1, -4, -2]
    ],
    [
        [9, -2, 3, 2],
        [2, 8, -2, 3],
        [-3, 2, 11, -4],
        [-2, 3, 2, 10]
    ]
]

Ys = [
    [41, -10, 1],
    [34, -32, 62],
    [55, -14, 12, -21]
]

# Proses setiap sistem persamaan
for idx, (A, b) in enumerate(zip(Xs, Ys), 1):
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    print(f"A = {A.tolist()}, y = {b.tolist()}")

    if check_diagonally_dominant(A):
        gauss_seidel(A, b)
    else:
        print("Melewati sistem ini karena tidak memenuhi syarat diagonal dominan.")
```

```python
# Newton Raphson E231-MATH6183-JJ01-04
import numpy as np

# Fungsi f(x)
def f(x):
    return x**6 + 2*x**2 - 3

# Turunan f(x)
def f_prime(x):
    return 6*x**5 + 4*x

# Metode Newton-Raphson
def newton_raphson(x0, tol=0.01, max_iter=15):
    # print(f"Initial guess: x0 = {x0}")

    for i in range(1, max_iter + 1):
        fx = f(x0)
        fpx = f_prime(x0)

        if fpx == 0:
            print("Error: Turunan nol. Tidak dapat lanjut iterasi.")
            return None

        x1 = x0 - fx / fpx
        print(f"Iteration {i}. new root = {x1}")

        if abs(x1 - x0) < tol:
            print(f"\nthe root of equations is: {x0}")
            return

        x0 = x1

    print("\nError: Melebihi jumlah maksimum iterasi. Root tidak ditemukan.")
    return None

# Panggil fungsi dengan tebakan awal 4
newton_raphson(x0=4, tol=0.01, max_iter=15)
```

```python
# Bisection E231-MATH6183-JJ01-04
# Solving Root Finding using Bisection Method (with numpy)
# Bisection E231-MATH6183-JJ01-04
import numpy as np

def f(x):
    return x**3 + 4*x**2 - 10

def bisection(a, b, x1, x2, tol=0.0001, max_iter=100):
    # Cek apakah interval valid
    if np.sign(f(a)) * np.sign(f(b)) >= 0:
        print(f"a = {a} and b = {b} not bound a root")

    # Tetap lanjutkan proses bisection
    for i in range(max_iter):
        c = (a + b) / 2
        fc = f(c)

        if np.abs(fc) < tol or (b - a) / 2 < tol:
            print(f"root of the function = {c} with coordinates of x1 = {x1} and x2 = {x2}")
            return

        if np.sign(f(a)) * np.sign(fc) < 0:
            b = c
        else:
            a = c

    print("Tidak konvergen.")

# Data kasus
cases = [
    (3, 4, -10, 17),       # Not bound
    (-2, 0, 10, -15),      # Valid
    (-3, 0, -2, 4),        # Valid
    (-2, 1, 12, -5),       # Valid
    (-8, -13, 7, -21)      # Not bound
]

for a, b, x1, x2 in cases:
    bisection(a, b, x1, x2)
```

```python
# Integral Rieman E231-MATH6183-JJ01-05
# Integral Riemann & Trapezoid Rule
import numpy as np

# Fungsi yang akan diintegrasikan
def f(x):
    return 4*x**4+2*x**3-6*x

# Interval integrasi
a = 0
b = 10

# ---------------------
# Riemann Sums (30 titik)
# ---------------------
n_riemann = 30
x_left = np.linspace(a, b, n_riemann + 1)[:-1]   # titik kiri
x_right = np.linspace(a, b, n_riemann + 1)[1:]   # titik kanan
x_mid = (x_left + x_right) / 2                   # titik tengah
dx_r = (b - a) / n_riemann

left_riemann = np.sum(f(x_left)) * dx_r
right_riemann = np.sum(f(x_right)) * dx_r
mid_riemann = np.sum(f(x_mid)) * dx_r


# ---------------------
# Output
# ---------------------
print(f"Left Riemann Sum     = {left_riemann}")
print(f"Right Riemann Sum    = {right_riemann}")
print(f"Midpoint Riemann Sum = {mid_riemann}")
```

```python
## Trapezoid Rieman E231-MATH6183-JJ01-05
import numpy as np

# Definisikan fungsi f(x)
def f(x):
    return np.sin(x)**2 + np.sin(x) + 2

# Batas integral
a = -np.pi
b = 2 * np.pi

# Jumlah titik grid
n = 27
x = np.linspace(a, b, n)
dx = (b - a) / (n - 1)

# Trapezoid Rule
trapezoid = (f(x[0]) + f(x[-1]) + 2 * np.sum(f(x[1:-1]))) * dx / 2

# Output
print(f"Hasil integral (Trapezoid Rule) ≈ {trapezoid}")
```
