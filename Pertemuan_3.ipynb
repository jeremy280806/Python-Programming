```python
daftarData = ["SMA N 1 SEMARANG", 11, 165.5, True, False, "seragam"]
print(daftarData) # sebelum
# start:stop:step
print(daftarData[0:3])
print(daftarData[:3])
print(daftarData[:3])
print(daftarData[3:])
print(daftarData[3:])
print(daftarData[1:6:2])
```

```python
# buatlah lsit yang mengandung = nama sekolah, kelas, dan rata-rata tinggi badan
list = ["SMA N 1 SEMARANG", 11, 165.5]
print(list) # sebelum di ubah
# ubah nama sekolah
list[0] = "SMA N 12 SEMARANG"
print(list) # setelah diubah
```

```python
# buatlah list yang mengandung = nama sekolah, kelas, dan rata-rata tinggi badan
list = ["SMA N 1 SEMARANG ", 11, 165.5]
print(list) # sebelum di ubah

#ubah nama sekolah
list[0] = "SMA N 12 SEMARANG"
print(list) # setelah diubah

#tambahkan gender, jumlah siswa laki-laki dalam satu kelas, dan rara-rata berat badan
list.append("laki-laki")
list.append(20)
list.append(50.0)
print(list)
```
