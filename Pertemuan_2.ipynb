```python
jenis_keanggotaan = str(input("Jenis keanggotaan: "))
total_pembelian = int(input("Total pembelian: "))

if jenis_keanggotaan == "member":
  if total_pembelian >= 1000000:
    diskon = 0.5
  if total_pembelian >= 500000:
    diskon = 0.1

else:
  if total_pembelian >= 1000000:
    diskon = 0.1
  else:
    diskon = 0

total_harga = total_pembelian - (total_pembelian * diskon)
print(f"Total harga Rp{total_harga}")
```
