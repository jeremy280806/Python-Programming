```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return x**2 + 3*x + 1
```

```python
def bisection(a, b, e = 0.01, n = 15):
  """
  Function untuk cari akar dengan bisection method
  a: tebakan swal untuk sisi kiri
  b: tebakan awal untuk sisi kanan
  e: toleransi error
  n: jumlah iterasi maksimal
  """
  # Cek hasil f(a) dan f(b) gaboleh punya tanda yang sama
  if np.sign(f(a)) == np.sign(f(b)):
    print("Tebakan tidak valid")
    return False
  # Cek apakah sudah sampai iterasi maksimal
  if(n <= 0):
    print("Jumlah iterasi maksimal tercapai dan hasil belum ditemukan!")
    return False

  # Cari c (nilai tengah dari a dan b)
  c = (a + b) / 2
  print(f"Mencoba nilai c: {c}, f(c): {f(c)}")

  # Cek apakah hasil sudah di bawah batas error
  if(np.abs(f(c)) <= e):
    print(f"Hasil ketemu di x + {c}")
    return True

  # Kalauhasil dari f(c) tandanya sama dengan f(a) maka kita geser "a"
  if(np.sign(f(c)) == np.sign(f(a))):
    return bisection(c, b, e, n - 1)
  else:
    # Kalau hasil dari f(c) tandanya sama dengan f(b) maka kita geser "b
    return bisection(a,c,e,n-1)
```

```python
bisection(-1, 2)
```

```python
def df(x):
  return 2*x + 3

def newton_raphson(x0, e = 0.01, n = 15):
  """
  Function untuk cari akar dengan bisection method
  x0: tebakan awal
  e: toleransi error
  n: jumlah iterasi maksimal
  """
  # Cek apakah sudah sampai iterasi maksimal
  if(n <= 0):
    print("Jumlah iterasi maksimal tercapai dan hasil belum ditemukan!")
    return False

  # Cari titik tangent
  tangent = f(x0) / df(x0)
  # Geser nilai x0 sejauh garis tangentnya
  x1 = x0 - tangent
  print(f"Mencoba nilai x1: {x1}, f(x1): {f(x1)}")
  return newton_raphson(x1, e, n - 1)
```
