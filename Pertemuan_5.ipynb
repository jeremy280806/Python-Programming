```python
# tuple
barang = str(input("Masukkan nama barang:"))
kategori_barang = str(input("Masukkan kategori barang (ELektronik/Pakaian/Makanan):"))
harga_barang = int(input("Masukkan harga barang:"))


iniTuple = ("Elektronik", "Pakaian", "Makanan")
if kategori_barang == "Elektronik":
  if harga_barang >= 0.10:
    diskon = 0.1
elif kategori_barang == "Pakaian":
  if harga_barang >= 0.20:
    diskon = 0.05
elif kategori_barang == "Makanan":
  if harga_barang >= 0:
    diskon = 0
else:
  diskon = 0


total_harga = harga_barang - (harga_barang * diskon)
print(f"Barang kabel dalam kategori {kategori_barang} mendapat diskon {}")
print(f"Harga setelah diskon: {total_harga}")
```
