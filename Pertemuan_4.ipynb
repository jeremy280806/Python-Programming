```python
# List
pengeluaran_satu = str(input("Masukkan Pengeluaran Pertama: "))
pengeluaran_dua = str(input("Masukkan Pengeluaran Kedua: "))
pengeluaran_tiga = str(input("Masukkan Pengeluaran Ketiga: "))

kategori = "boros", "cukup", "hemat"
list = [pengeluaran_satu, pengeluaran_dua, pengeluaran_tiga]
if kategori == "boros":
  if pengeluaran_satu >= 20000000:
    print(f"Pengeluaran pertama sebesar {pengeluaran_satu} termasuk kategori boros")
  elif pengeluaran_dua >= 20000000:
    print(f"Pengeluaran kedua sebesar {pengeluaran_dua} termasuk kategori boros")
  elif pengeluaran_tiga >= 20000000:
    print(f"Pengeluaran ketiga sebesar {pengeluaran_tiga} termasuk kategori boros")
  else:
    print("Tidak ada yang termasuk kategori boros")

if kategori == "cukup":
  if pengeluaran_satu >= 5000000:
    print(f"Pengeluaran pertama sebesar {pengeluaran_satu} termasuk kategori cukup")
  elif pengeluaran_dua >= 5000000:
    print(f"Pengeluaran kedua sebesar {pengeluaran_dua} termasuk kategori cukup")
  elif pengeluaran_tiga >= 5000000:
    print(f"Pengeluaran ketiga sebesar {pengeluaran_tiga} termasuk kategori cukup")
  else:
    print("Tidak ada yang termasuk kategori cukup")

if kategori == "hemat":
  if pengeluaran_satu <= 5000000:
    print(f"Pengeluaran pertama sebesar {pengeluaran_satu} termasuk kategori hemat")
  elif pengeluaran_dua <= 5000000:
    print(
        f"Pengeluaran kedua sebesar {pengeluaran_dua} termasuk kategori hemat")
  elif pengeluaran_tiga <= 5000000:
    print(f"Pengeluaran ketiga sebesar {pengeluaran_tiga} termasuk kategori hemat")
  else:
    print("Tidak ada yang termasuk kategori hemat")

print(f"Pengeluaran Pertama: {pengeluaran_satu} termasuk kategori {kategori[0]}")
print(f"Pengeluaran Kedua: {pengeluaran_dua} termasuk kategori {kategori[1]}")
print(f"Pengeluaran Ketiga: {pengeluaran_tiga} termasuk kategori {kategori[2]}")
```

```python
key = 1234
percobaan = 0
max_percobaan = 3
while percobaan < max_percobaan:
  pin = int(input("Masukkan PIN: "))
  if pin == key:
    print("Selamat datang di BANK XYZ!")
    break
  else:
    percobaan += 1
    print("PIN ANDA SALAH")
  if percobaan == max_percobaan:
    print("TERBLOKIR")
```
