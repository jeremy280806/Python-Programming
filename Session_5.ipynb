```python
import numpy as np
```

```python

# Left Riemann
def f(x):
  return x**2

def g(x):
  return x**3/3


# Start
a = 0
# End
b = 4

# Mau berapa persegi panjang (semakin banyak semakin akurat)
n = 100

# Tentukan lebar setiap pesergi panjang
width = (b-a)/n

# Dapatkan hasil asli
hasil_asli = g(b)-g(a)
print(f"Hasil asli: {hasil_asli}")

# Left Rieman
# Cukup ambil titik kirinya
x = np.linspace(a, b-width, n)
y = f(x)
print(x)

# Tinggal hitung hasil
# Terus dijumlah
hasil = np.sum(width * y)
print(f"Hasil left Riemann: {hasil}")
print(f"Error: {hasil_asli - hasil}")
```

```python
# Right Riemann
def f(x):
  return x**2

def g(x):
  return x**3/3


# Start
a = 0
# End
b = 4

# Mau berapa persegi panjang (semakin banyak semakin akurat)
n = 100

# Tentukan lebar setiap pesergi panjang
width = (b-a)/n

# Dapatkan hasil asli
hasil_asli = g(b) - g(a)
print(f"Hasil asli: {hasil_asli}")

# Right Rieman
# Cukup ambil titik kirinya
x = np.linspace(a + width, b, n)
y = f(x)
print(x)

# Tinggal hitung hasil
# Terus dijumlah
hasil = np.sum(width * y)
print(f"Hasil left Riemann: {hasil}")
print(f"Error: {hasil_asli - hasil}")
```

```python
# Midpoint Riemann
# Cukup ambil titik tengahnya
x = np.linspace (a + width/2, b - width /2, n)
y = f(x)
print(x)
# Tinggal hitung hasil pakai rumus persegi panjang: p x l
# Terus dijumlahkan
hasil = np.sum(width * y)
print(f"Hasil  Midpoint Riemann: {hasil}")
print(f"Error: {hasil_asli - hasil}")
```

```python
# Trapezoid Riemann
# Cukup ambil titik tengahnya
x = np.linspace (a, b, n+1)
y = f(x)
print(x)
# Tinggal hitung hasil pakai rumus trapesium: (a+b)*t/2
# Terus dijumlah
hasil = np.sum(y[1:] + y[:-1]) * width / 2
print(f"Hasil  Midpoint Riemann: {hasil}")
print(f"Error: {hasil_asli - hasil}")
```
