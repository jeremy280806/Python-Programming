```python
# study case: Mencari nama dan Nilai Pemenang Olimpiade
data_peserta = {}
jumlah_peserta = int(input("Masukkan Jumlah Peserta : "))
for i in range(1, jumlah_peserta + 1):
  nama_peserta = input(f"Peserta ke-{i}: ")
  nilai_peserta = int(input(f"Nilai {nama_peserta}: "))
  data_peserta[nama_peserta] = nilai_peserta

print("\nData nilai siswa: ")
for nama, nilai in data_peserta.items():
  print(f"{nama}: {nilai}")

nilai_tertinggi = max(data_peserta.values())
for nama, nilai in data_peserta.items():
  if nilai == nilai_tertinggi:
    print(f"\nPemenang olimpiade adalah {nama} dengan nilai {nilai}")
```

```python
# hitung pembayaran perminggu dengan time - dan - setengah untuk overtime
def bayar(gaji,hours):
  if hours <= 40:
    amount = gaji * hours
  else:
    amount = (gaji * 40) + ((1.5 * gaji) * (hours - 40))
  return amount
print(bayar(20000,41))
```

```python
# Study Case: Mencari Nilai Setiap Pelajaran dari Peserta Olimpiade

# Letakkan data-data nama dan nilai peserta olimpiade ke dalam dictionary bersarang
dataPeserta = {
    "Gaga": {"Matematika": 85, "Bahasa Inggris": 90, "IPA": 88},
    "Lia": {"Matematika": 78, "Bahasa Inggris": 82, "IPA": 80},
    "Rayhan": {"Matematika": 92, "Bahasa Inggris": 80, "IPA": 95}
}

# mintalah user untuk menginput nama peserta olimpiade
namaPeserta = input("Masukkan nama siswa: ").capitalize()
# periksa adanya nama di dictionary
if namaPeserta in dataPeserta:# jika <variabel input (namaPeserta)> in data Peserta
    print(f"Nilai {namaPeserta}:")
    for pelajaran, nilai in dataPeserta[namaPeserta].items():# untuk pelajaran(value), nilai dalam dataPeserta[key].item()
        print(f"{pelajaran}: {nilai}")
else:
  print("Nama peserta tidak ditemukan.")
```

```python
# Latihan Menghitung Jumlah Kembalian di Kasir

def hitung_kembalian(total_uang):
  if total_uang < 1000:
    print(f"Tidak ada kembalian: {total_uang}")
    return 0
  else:
    print("Uang kembali: 1 lembar Rp1000")
    return 1+hitung_kembalian(total_uang-1000)

uang = int(input("Masukkan total uang kembalian: "))
lembar = hitung_kembalian(uang)
print(f"Total lembar uang yang diberikan: {lembar}")
```
