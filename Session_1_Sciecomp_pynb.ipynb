```python
#conditional / if-else
#Contoh kasus:  bilangan posive vs negative vs nol & genap vs ganjil
angka = Int(input("Masukkan angka"))
if angka > 0:
    if angka % 2 == 0:
      print("Bilangan positif genap")
      print("Saya suka saya suka")
    else:
      print("Bilangan positif ganjil")
elif angka < 0:
  if angka % 2 == 0:
    print("Bilangan negatif genap")
  else:
    print("Bilangan negatif ganjil")
else:
  print("Nol")
```

```python
myList = ["James", 3, "Yanto", True, 3.14]
print(myList[2])
print(myList[-1])
print(myList[-3])
```

```python
#Multiple Indexing
#Syntax: <List_name>[start : end : step]
#End is exclusive
#Kalau ada yang ga diisi bakal pakai default
#Default start: value pertama
#Default end: setelah value terakhir
#Step: 1

print(myList[1:4])
print(myList[1:4:2])
print(myList[2::2])

```
